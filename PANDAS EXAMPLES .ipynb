```python
# Importing necessary libraries
import pandas as pd
import numpy as np

# Creating a DataFrame using a dictionary
data_dict = {
    'A': np.random.rand(5),
    'B': np.random.randint(1, 100, 5),
    'C': np.random.choice(['X', 'Y', 'Z'], 5)
}
df_from_dict = pd.DataFrame(data_dict)
print("DataFrame created from dictionary:")
print(df_from_dict)

# Creating a DataFrame using a random list of lists
data_list = [
    [np.random.rand() for _ in range(3)],
    [np.random.randint(1, 100) for _ in range(3)],
    np.random.choice(['X', 'Y', 'Z'], 3)
]
df_from_list = pd.DataFrame(data_list, index=['Row1', 'Row2', 'Row3'], columns=['A', 'B', 'C'])
print("\nDataFrame created from list of lists:")
print(df_from_list)

# Creating a Series from a random list
random_series_list = np.random.randn(5)  # Generating a random list
series_from_list = pd.Series(random_series_list, name='Random Series')
print("\nSeries created from a random list:")
print(series_from_list)

# Creating a Series from a dictionary
random_dict = {'A': np.random.rand(), 'B': np.random.rand(), 'C': np.random.rand()}
series_from_dict = pd.Series(random_dict)
print("\nSeries created from a dictionary:")
print(series_from_dict)
```

```python
# Importing necessary libraries
import pandas as pd

# Creating a DataFrame with some missing values and duplicate rows
data = {
    'A': [1, 2, None, 4, 5],
    'B': [6, None, 8, 8, 10],
    'C': ['a', 'b', 'c', 'd', 'e']
}
df_dirty = pd.DataFrame(data)
df_dirty = pd.concat([df_dirty, df_dirty])  # Adding duplicate rows
print("DataFrame with missing values and duplicate rows:")
print(df_dirty)

# Data Cleaning

# Removing duplicate rows
df_cleaned = df_dirty.drop_duplicates()
print("\nDataFrame after removing duplicate rows:")
print(df_cleaned)

# Filling missing values with a specific value
df_cleaned['A'] = df_cleaned['A'].fillna(0)  # Filling missing values in column 'A' with 0
df_cleaned['B'] = df_cleaned['B'].fillna(method='ffill')  # Filling missing values in column 'B' with forward fill
print("\nDataFrame after filling missing values:")
print(df_cleaned)

# Dropping rows with missing values
df_cleaned = df_cleaned.dropna()
print("\nDataFrame after dropping rows with missing values:")
print(df_cleaned)
```

```python
# Importing necessary libraries
import pandas as pd

# Creating two sample DataFrames
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}
Df1 = pd.DataFrame(data1)

data2 = {
    'ID': [3, 4, 5, 6],
    'Department': ['HR', 'Finance', 'Marketing', 'IT'],
    'Salary': [5000, 6000, 5500, 7000]
}
Df2 = pd.DataFrame(data2)

# Merging DataFrames based on a common column (ID)
merged_df = pd.merge(Df1, Df2, on='ID', how='inner')  # Inner merge
print("\nMerged DataFrame based on ID:")
print(merged_df)

# Concatenating DataFrames along rows
concatenated_df = pd.concat([Df1, Df2], ignore_index=True)
print("\nConcatenated DataFrame along rows:")
print(concatenated_df)
```

```python


def convert_subs_to_int(subs_type):
    if subs_type == 'Basic':
        return 1
    elif subs_type == 'Premium':
        return 2
    elif subs_type == 'Standard':
        return 3


# Applying the function to the 'Subscription Type' column
df['Subscription Type'] = df['Subscription Type'].apply(convert_subs_to_int)
df
```
